```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('weather_data.csv')
df
```

```python
df1 = df[['name','precip','windspeed','conditions']]
```

```python
df1
```

```python
df1['precip'].nlargest(n=40)
```

```python
df1['windspeed'].nlargest(n=25)
```

```python
df1['Wind_Cond'] = np.where(df['windspeed']>30,True,False)
```

```python
df1['Precip_Cond'] = np.where(df['precip']>1,True,False)
```

```python
df1
```

```python
df1[df1['Wind_Cond']==True]
```

```python
df1[df1['Precip_Cond']==True]
```

```python
df1
```

```python
conditions = [
    (df1['Wind_Cond'] == True) & (df1['Precip_Cond'] == False),
    (df1['Precip_Cond'] == True)  & (df1['Wind_Cond'] == False),
    (df1['Wind_Cond'] == True) &  (df1['Precip_Cond'] == True),
    (df1['Wind_Cond'] == False) &  (df1['Precip_Cond'] == False)
    ]

values = [1,2,3,4]
df1['tier'] = np.select(conditions, values)
```
