```python
import xml.dom.minidom as xmlreader
import pandas as pd

doc = xmlreader.parse("export.osm")

print(doc.firstChild.tagName)
```

```python
actions = [] # keep only "modify"
old_or_new = []
street_id = []
street_name = []
maxspeed = []
```

```python
acts = doc.getElementsByTagName("action")

for act in acts: 
    if act.getAttribute("type") != "modify": continue
    actions.append(act.getAttribute("type"))

ways = doc.getElementsByTagName("way")

for way in ways:
    street_id.append(way.getAttribute("id"))

oldnews = doc.getElementsByTagName("old")
```

```python
import xml.etree.ElementTree as ET

tree = ET.parse('export.osm')
root = tree.getroot()

full_list = []
for act in root: # action, tag: action, attrib: type = "XX"
    for key, value in act.attrib.items(): 
        if value != "modify": continue
        actions.append(value) # all "modify"
        
        old_or_new = []
        for oldornew in act: # old & new
            street_id = []
            maxspeed = []
            street_name = []
            
            for way in oldornew: # way
                for key, value in way.attrib.items():
                    if key == "id": street_id.append(value)
                    for tags in way: # nd & tags
                        if "k" in tags.attrib.keys() and tags.attrib["k"] == 'maxspeed':            
                            maxspeed.append(tags.attrib["v"])
                        if "k" in tags.attrib.keys() and tags.attrib["k"] == 'name':            
                            street_name.append(tags.attrib["v"])
            if len(street_name) > 0:
                old_or_new.append({'id':street_id[0], 'tag':oldornew.tag, 'name':street_name[0], 'maxspeed':maxspeed[0]})
        if len(old_or_new) > 1:
            full_list.extend(old_or_new)
full_list
df = pd.DataFrame(full_list)
df
```

```python
#df = (df
#        .groupby('id').agg({})
#        .reset_index())
diff = df.pivot(index='id', columns='tag', values='maxspeed')
diff = diff[diff['new'] != diff['old']]
diff
```
