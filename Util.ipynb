```python
import pandas as pd
import os
```

```python
root = 'bikecounter_data/'
#for file in os.listdir(root):
#    df = pd.read_csv('bikes.csv') root+file 
all_files = [root+file for file in os.listdir(root)]
df = pd.concat((pd.read_csv(f) for f in all_files), ignore_index=True)
df.to_csv('bike_counters_full.csv')
```

```python
root = 'strava_data_counters/'
#for file in os.listdir(root):
#    df = pd.read_csv('bikes.csv') root+file 
#all_files = [root+file for file in os.listdir(root)]
all_files = []
filetype = '.csv'
for (dir_path, dir_names, files) in os.walk(root):
    for file in files:
         if file.lower().endswith(filetype.lower()):
            all_files.append(os.path.join(dir_path, file))
df = pd.concat((pd.read_csv(f) for f in all_files), ignore_index=True)
df.to_csv('strava_counters_year.csv')
df
```
